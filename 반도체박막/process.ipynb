```python
import pandas as pd
train = pd.read_csv("trains.csv")
train.head()
```

```python
train.head()
```

```python
test = pd.read_csv("test.csv")
test.drop("id",axis=1,inplace=True)
test.head()
```

```python
df = pd.concat([train,test],axis=0)
df.shape
```

```python
import numpy as np

for col in df.columns:
    col_type = df[col].dtypes
    min1 = df[col].min()
    max1 = df[col].max()
    if str(col_type)[:3] == 'int':
        df[col] = df[col].astype(np.int16)
    else:
        if min1 > np.finfo(np.float16).min and max1 < np.finfo(np.float16).max:
            df[col] = df[col].astype(np.float16)
        elif min1 > np.finfo(np.float32).min and max1 < np.finfo(np.float32).max:
            df[col] = df[col].astype(np.float32)
        else:
            df[col] = df[col].astype(np.float64)
```

```python
df.info()
```

```python
train_x = train.iloc[:,:226]
train_y = train.iloc[:,226:]
train_x.head()
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
x_train_sca = scaler.fit_transform(train_x)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(train_x,train_y, test_size=0.2, random_state=0)
```

```python
train_x.corr()
```

```python
import lightgbm as lgb
train_set = lgb.Dataset(x_train, y_train["layer_1"], silent=False)
valid_set = lgb.Dataset(x_val, y_val["layer_1"], silent=False)

#lgb_dtrain = lgb.Dataset(data = pd.DataFrame(train_x), label = pd.DataFrame(train_y["layer_1"])) # 학습 데이터를 LightGBM 모델에 맞게 변환
lgb_param = {'boosting_type':'gbdt',
             'num_leaves': 30,
             'max_depth': -1,
            'learning_rate': 0.01, 
            'bagging_fraction' : 1,
            'max_bin' : 5000 ,
            'bagging_freq': 20,
            'colsample_bytree': 0.6,
             'metric': 'mae',
             'min_split_gain': 0.5,
            'min_child_weight': 1,
            'min_child_samples': 10,
            'scale_pos_weight':1,
            'zero_as_missing': True,
            'objective': 'regression', # 목적 함수
            }
lgb_model = lgb.train(params = lgb_param, train_set = train_set , num_boost_round=500, early_stopping_rounds=50,verbose_eval=50, valid_sets=valid_set)
#lgb_model_predict = np.argmax(lgb_model.predict(X_test), axis = 1)
```
